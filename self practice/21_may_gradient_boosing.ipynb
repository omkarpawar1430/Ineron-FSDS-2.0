```python
import numpy as np
import pandas as pd 
```

```python
df = pd.DataFrame({'iq':[90, 100, 110, 120, 80], 'cgpa': [8, 7, 6, 9, 5], 'salary': [3, 4, 8, 6, 3]})
df
```

```python
df['base_model'] = np.mean(df['salary'])
```

```python
df
```

```python
df['residual_1'] = df['salary'] - df['base_model']
df
```

```python
X = df[['iq', 'cgpa']]
```

```python
y = df['residual_1'] # instead of salary as our Target variable we are taking residual of salary. 
```

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
```

```python
model.fit(X, y)
```

```python
model.score(X, y)
```

```python
# plotting of the decision Tree:

from sklearn import tree
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
tree.plot_tree(model,filled=True)
plt.show()
```
