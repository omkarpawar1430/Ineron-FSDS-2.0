### Q 6 | Ans:

Make an OrderedDict called fancy from the same pairs listed in 5 and print it. Did it print in the
same order as plain?

```python
from collections import OrderedDict

plain_dict = {'a': 1, 'c': 3, 'b': 2}
fancy = OrderedDict(plain_dict)

print(fancy)
```

Normal dictionary don't preserve order of the keys: values as they are added. When we use Ordered Dictionary we can preserve order of the keys: values.

### Q 7 | Ans:

Make a default dictionary called dict_of_lists and pass it the argument list. Make the list
dict_of_lists[&#39;a&#39;] and append the value &#39;something for a&#39; to it in one assignment. Print
dict_of_lists[&#39;a&#39;].

```python
from collections import defaultdict

dict_of_lists = defaultdict(list)
dict_of_lists['a'].append('something for a')
dict_of_lists['b'].append('something for b')

print(dict_of_lists)

print(dict_of_lists['a'])
```
