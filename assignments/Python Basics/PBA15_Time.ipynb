### Q 7 | Ans: 

```python
def genPrimes():
    yield 2 
    primes = [2] 
    n = 3 
    while True:
        for prime in primes:
            if n % prime == 0:
                break 
        else:
            yield n
            primes.append(n)
        n += 2 
gen = genPrimes()
```
