Que1. create a file class for reading a data from one of the file with a method name read and write.

note: Use logging and exception handling

```python
# import modules for logging and create logger:
import logging
logging.basicConfig(filename="file_class.log", filemode="w", format="%(levelname)s %(asctime)s %(message)s")
logger = logging.getLogger()
logger.setLevel(logging.DEBUG)


# Create file class
class file():
    def __init__(self, filename):
        self.filename = filename
        logger.info(f"working on file {self.filename}")
    def read(self):
        logger.info(f"Reading file {self.filename}")
        try:
            with open(self.filename, "r") as f:
                result = f.read()
            return result
        except Exception as e:
            logger.error(f"File does not exist {self.filename}, look if you have misspelled file name!")
    
    def write(self,msg):
        try:
            with open(self.filename, "a") as f:
                f.write(msg)
                logger.info(f"{msg} written successfully!")
        except Exception as e:
            logger.error(f"Error: {e}")
```

### Q.1: Create your own class to perform a list search operation 

```python
# Import modules
import logging
# Create logger
logging.basicConfig(filename="que1.log", filemode="w", format="%(levelname)s %(asctime)s %(message)s")
logger = logging.getLogger()
logger.setLevel(logging.DEBUG)

# Create Class(Tip: Use exception handling + logging)
class List_Search:
    def __init__(self, list1):
        self.list1 = list1
    def compare(self, item):
        logger.info(f"Looking for {item} in {self.list1}")
        try:
            for i in self.list1:
                logger.debug(f"checking if {i} == {item}")
                if i == item:
                    logger.info(f"{item} exists in {self.list1}")
                    return f"{item} exists in {self.list1}"
            else:
                logger.warning(f"{item} does not exist in {self.list1}")
                return f"{item} does not exist in {self.list1}"
        except Exception as e:
            logger.error(f"Error: {e}")


# Create Object
num = [1, 2, 3, 5, 'ai']
ls_obj = List_Search(num)
# Call Methods of Class using object constructor
ls_obj.compare(3)
```
