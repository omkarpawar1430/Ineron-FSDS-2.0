# Data Visualization

### Bar Graph: 


#### sns.barplot()

A bar graph is most useful when you want to compare the quantities or frequencies of different categories. Here are some specific scenarios where using a bar graph is particularly beneficial:

1. **Comparing Categorical Data**: Bar graphs are excellent for comparing the values of different categories or groups. For instance, you might want to compare sales performance across different products, student performance in different subjects, or customer satisfaction levels for various services.

2. **Showing Distribution**: When you have discrete categories and you want to illustrate the distribution or frequency of each category, a bar graph is a clear and effective choice. It provides a visual representation of how often each category occurs.

3. **Highlighting Differences**: Bar graphs make it easy to highlight and compare differences in the quantities of categories. This is useful when you want to emphasize variations in data points.

4. **Visualizing Ranking or Ordering**: If your categories have a natural order or ranking, a bar graph can effectively convey this information. The length of the bars provides a straightforward visual representation of the relative magnitudes.

5. **Presenting Nominal or Ordinal Data**: Bar graphs are suitable for both nominal and ordinal data. Nominal data has categories with no inherent order, while ordinal data has categories with a meaningful order.

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Creating a mini dataset
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Count': np.random.randint(1, 20, 5)  # Generating random counts for each category
}

df = pd.DataFrame(data)

# Creating a bar plot using Seaborn
sns.barplot(x='Category', y='Count', data=df, palette='viridis')

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Bar Plot for Categorical Data')

# Display the plot
plt.show()
```

#### sns.barplot() vs sns.countplot()

Use `barplot` when:

1. You have numerical data associated with each category.
2. You want to visualize the central tendency or distribution of numerical data.
3. Aggregating numerical values (e.g., mean, median) for each category is essential.

Use `countplot` when:

1. Your primary interest is visualizing the frequency or count of occurrences for each category.
2. Exploring class imbalances in classification problems is needed.
3. You're conducting a simple exploration of categorical data without aggregating numerical values.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a dataset
data = {
    'Category': ['A', 'B', 'C', 'A', 'B', 'A', 'C', 'C', 'C'],
}

df = pd.DataFrame(data)

# Using countplot
sns.countplot(x='Category', data=df)
plt.title('Seaborn countplot')

plt.show()
```

### Histogram

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a seed for reproducibility
np.random.seed(42)

# Generating synthetic data for 'Age'
num_samples = 500
mean_age = 25
std_dev_age = 20

# Creating a DataFrame
data = {
    'Age': np.random.normal(loc=mean_age, scale=std_dev_age, size=num_samples)
}

df = pd.DataFrame(data)

# Creating a histogram
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue')

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age')

# Display the plot

plt.show()
```

### stacked histogram 

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset from Seaborn
tips = sns.load_dataset("tips")

# Creating a stacked histogram using Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(data=tips, x='total_bill', hue='day', multiple='stack', palette='pastel', edgecolor='w', bins=30)

# Adding labels and title
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.title('Stacked Histogram of Total Bill Across Days')

# Displaying the plot
plt.show()
```

When to Use a Histogram:
1. Data Distribution: Histograms are ideal for visualizing the distribution of a dataset. They show the frequency of values within different bins.
2. Skewness and Outliers: Histograms help identify skewness in the data and potential outliers.
3. Choosing Bin Sizes: Histograms assist in determining appropriate bin sizes for data analysis.
4. Comparisons: They are useful when comparing the distributions of multiple variables or groups.

### Boxplot:

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a mini-dataset of 1000 random values for the box plot
np.random.seed(42)
data_box = np.random.normal(loc=0, scale=1, size=1000)

# Creating a box plot using Seaborn
sns.boxplot(x=data_box, color='lightblue')

# Adding labels and title
plt.xlabel('Values')
plt.title('Box Plot of Random Data')

# Displaying the plot
plt.show()
```

### Pie Chart

```python
import matplotlib.pyplot as plt

# Sample data for a pie chart
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [25, 30, 20, 25]  # Percentages should sum up to 100

# Creating a pie chart using Matplotlib
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightblue', 'lightgreen', 'lightskyblue'])

# Adding a title
plt.title('Distribution of Categories')

# Displaying the plot
plt.show()
```

```python
import pandas as pd

# Corrected sample DataFrame
data = {
    'Category': ['A', 'B', 'C', 'D', 'A', 'A', 'B', 'C', 'D', 'D', 'D', 'E']
}

df = pd.DataFrame(data)

import matplotlib.pyplot as plt

# Creating a pie chart from a pandas DataFrame
plt.pie(df['Category'].value_counts(), labels=df['Category'].unique(), autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightblue', 'lightgreen', 'lightskyblue', 'lightpink'])

# Adding a title
plt.title('Distribution of Categories')

# Displaying the plot
plt.show()

```

seaborn don't have special pie chart functionality. we can use plt.pie()

When to Use a Pie Chart:
1. Proportional Representation: Pie charts are effective when you want to represent the proportional distribution of categories in a whole.
2. Limited Categories: They work well when you have a small number of categories (3-7) to avoid visual clutter.
3. Highlighting a Part-to-Whole Relationship: Useful for emphasizing the contribution of each category to the whole.

----------

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Sample DataFrame
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016],
    'Value': [15, 18, 22, 28, 35, 40, 45]
}

df = pd.DataFrame(data)

# Creating a line graph using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Value', 
             data=df, marker='o', 
             color='skyblue', 
             linestyle='-', 
             markersize=20)

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Line Graph of Values Over Time')

# Displaying the plot
plt.show()
```

When to Use a Line Graph:
1. Trends Over Time: Line graphs are excellent for showing trends and patterns in data over time.
2. Sequential Data: Useful for displaying data points in sequential order, such as time series data.
3. Comparisons: Effective for comparing trends between different categories or groups.
4. Highlighting Changes: Helps in highlighting changes or fluctuations in values.

### Scatter plot:

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset from Seaborn
tips = sns.load_dataset("tips")

# Display the first few rows of the dataset
print(tips.head())

# Creating a scatter plot using Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips, color='skyblue', marker='o', s=60)

# Adding labels and title
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.title('Scatter Plot of Total Bill vs Tip')

# Displaying the plot
plt.show()
```

Key points for using scatter plots:

1. Relationship Exploration: Scatter plots are effective for visually exploring relationships between two continuous variables.
2. Identification of Trends: Patterns in the scatter plot can reveal trends, clusters, or outliers in the data.
3. Correlation Assessment: The overall pattern and direction of points can provide insights into the correlation between variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the "tips" dataset from Seaborn
tips = sns.load_dataset("tips")

# Creating a violin plot using Seaborn
plt.figure(figsize=(10, 6))
sns.violinplot(x='day', y='total_bill', data=tips, palette='pastel')

# Adding labels and title
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.title('Violin Plot of Total Bill Across Days')

# Displaying the plot
plt.show()
```

Key points for using violin plots:

1. Distribution Exploration: Violin plots provide a summary of the distribution of the data, including information about the median, quartiles, and density.
2. Comparison Between Categories: Useful for comparing distributions across different categories.
3. Multivariate Analysis: Violin plots can be used to explore relationships between multiple variables.

### density plot: kdeplot()

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data (replace this with your dataset)
data = sns.load_dataset("iris")

# Create a density plot using Seaborn
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))

# Adjust the parameters as needed
sns.kdeplot(data['sepal_width'], fill=True, color='skyblue', label='Sepal Width')
sns.kdeplot(data['sepal_length'], fill=True, color='orange', label='Sepal Length')

# Set labels and title
plt.xlabel('Feature Values')
plt.ylabel('Density')
plt.title('Density Plot of Sepal Width and Length')

# Add legend
plt.legend()

# Show the plot
plt.show()
```
