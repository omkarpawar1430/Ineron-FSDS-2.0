```python
## Drop NaN values:

sub_t = t.iloc[0:6, [0,1,2,3,9,10]]
sub_t.dropna() # It removes NaN values works Row wish (axis = 0) but do not do permanent changes. to 
                # so you can reassign or (inplace=True)
```

```python
sub_t.dropna(axis=1) # removed column where null values are present
```

```python
sub_t
```

```python
sub_t.dropna(axis=1, thresh=2) # Observe Cabin Column, There are two Non NaN values are present
```

```python
sub_t.dropna(axis=1, thresh=3) # Cabin crosses threshold point and gets dropped
```
