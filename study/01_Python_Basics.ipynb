# Variables | Data Type | Numbers

## Variables:

1. Variables are containers for storing data values.
2. A variable is created the moment you first assign a value to it.
3. Variables do not need to be declared with any particular type, and can even change type after they have been set.

### Rules for choosing variable name:

1. A variable name must start with a letter or the underscore character
2. A variable name cannot start with a number
3. A variable name can only contain alpha-numeric characters and underscores (A-z, 0-9, and _ )
4. Variable names are case-sensitive (age, Age and AGE are three different variables)

````python
x = 125
print(x)
````

````python
name = "Omkar"
Name = "Snow"
# In this case, name and Name are both different variables 
# as python is case sensitive language.
print(name)
print(Name)
````

````python
name1 = "Rahul" # number after letters or underscore is allowed.
print(name1)
````

````python

1name = "omkar" # we can not start variable name with number but we can write it after letter or underscore.
print(1name)
````

### Multiple Value Assignment:

1. Python allows you to assign values to multiple variables in one line.
2. Make sure the number of variables matches the number of values, or else you will get an error.
3. you can assign the same value to multiple variables in one line.

Unpack a Collection:
If you have a collection of values in a list, tuple etc. Python allows you to extract the values into variables. This is called unpacking.


````python
x, y, z = 5, "omkar", True
print(x)
print(y)
print(z)
````

````python
# same value multiple variables:
x = y = z = "Orange"
print(x)
print(y)
print(z)
````

````python
#unpacking:
fruits = ["apple", "banana", "cherry"]
x, y, z = fruits
print(x)
print(y)
print(z)
````

# Casting 

Casting in python is therefore done using constructor functions:

1. int() - constructs an integer number from an integer literal, a float literal (by removing all decimals), or a string literal (providing the string represents a whole number)
2. float() - constructs a float number from an integer literal, a float literal or a string literal (providing the string represents a float or an integer)
3. str() - constructs a string from a wide variety of data types, including strings, integer literals and float literals

````python
x = input("enter a number: ")
print(type(x))

x = int(float(x)) # here we are type casting str to int.
"""
If the string represents a number with a decimal point,
you can use the float() function to convert it to a floating-point number,
or you can remove the decimal point and any digits after 
it to convert it to an integer.
"""   
print(type(x))
print(int(54.54)) # removes decimal points.
print(float(54))
````

## Slicing Strings

````python
a = "Omkar Dipak Pawar"
print(a[0:5]) # indexing starts with 0, 1, 2, ...
print(a[6:11])
print(a[12:17])
````

````python
a = "Omkar Dipak Pawar"
# slicing from the start :
print(a[:5]) # By default it will start from index 0
# slice to the end :
print(a[12:]) # By default it will slice to the end
````

````python
# Negative Slicing:
a = "Omkar Dipak Pawar"
print(a[-5:])
print(a[-1])
# Negative Slicing can be used as an alternative way. 
# ..., -3, -2, -1
````

````python
# Negative indexing:
print(a[::-1])
print(a[-5:-10:-1]) #! start:end:step 
````

## String Methods

### Converting String:

````python
name = "oMkar diPak PAwar"
print(name.capitalize())
print(name.title())
print(name.lower())
print(name.casefold()) # lower case and casefold are similar in nature.
print(name.upper())
print(name.swapcase())
````

### Measuring String:

````python
# count()	Returns the number of times a specified value occurs in a string
str = "I love to learn new things every day!"
print(str.count("e")) # gives count of times a string occurs in a string. 
````

### Identify in Strings:

````python
# The find() method finds the first occurrence of the specified value.
str = "I love to learn new things every day!"
print(str.find("n")) # output: 14
str.find("z") # output: -1 ( when value is not present in the string returns -1)
````

````python
# index() is similar to find(), but index() method raises an exception if the value is not found.
print(str.index("n")) # output: 14
str.index("z")
````

````python
# startswith()	Returns true if the string starts with the specified value
name = "Omkar"
result = name.startswith("O")
print(result)
# endswith()	Returns true if the string ends with the specified value
result = name.endswith("r")
print(result)
````

### checking type of string:

````python
# The isalnum() method returns True if all the characters are alphanumeric,
#  meaning alphabet letter (a-z) and numbers (0-9).
s1 = "pawaromkar1430"
result = s1.isalnum()
print(result)
````

````python
# The isalpha() method returns True if all the characters are alphabet letters (a-z).
s2 = "Omkar"
result = s2.isalpha()
print(result)
````

````python
# The islower() method returns True if all the characters are in lower case, otherwise False.
s3 = "omkar"
result = s3.islower()
print(result)
# isupper() returns True if all the characters are in upper case, otherwise False.
s4 = "PAWAR" # ! all characters should be upper case.
result = s4.isupper()
print(result)
# istitle()	Returns True if the string follows the rules of a title.add()
s5 = "Omkar Pawar" # First letter of the each character in the string should be 
s6 = "OMKAR"
print(s5.istitle())
print(s6.istitle())
````

````python
# The isdigit() method returns True if all the characters are digits, otherwise False.
txt = "938723"
result = txt.isdigit()
print(result)
````

````python
# isidentifier() method returns True if the string is a valid identifier else False.
a = "MyFolder"
b = "Demo002"
c = "2bring"
d = "my demo"

print(a.isidentifier())
print(b.isidentifier())
print(c.isidentifier())
print(d.isidentifier())
````

delimiter is a character or a sequence of characters used to separate or mark the boundaries between separate pieces of text or data.

In many cases, delimiters are used to separate individual elements in a larger data structure, such as a list or a string. For example, in a comma-separated values (CSV) file, the comma is used as a delimiter to separate different values in each row of the file.

````python
txt = "Lets try to split this line and create a list out of it."
print(txt.split()) # by default split by using whitespace as a separator.
txt = "lets split by using a special separator, shall we?"
print(txt.split(",")) # uses provided separator to split.
````
