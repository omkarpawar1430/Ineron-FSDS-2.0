```python
d1 = {'key1':[1,2,3,4,5,"Omkar"],       
      'key2' : [7,8,9,10,11,12],
      'key3':[13,14,15,16,17,18],
      'key4':[19,20,21,22,23,24]
}
df1 = pd.DataFrame(d1) # gives csv data into tabular format , easy to read
df1
```

```python
d2 = {'key1':[1,"Shiva",3,4,5,"Omkar"],       
      'key2' : [7,8,9,10,11,12],
      'key3':[13,"Data Sci",15,"AI",17,18],
      'key4':[19,20,21,22,23,24]
}
df2 = pd.DataFrame(d2)
df2
```

### Merge: 

```python
df1
```

```python
df2
```

Merges using intersection: (by default):
keeps data which is common in both of the data frames df1 and df2



```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2, how='inner', on='key3')
```

```python
pd.merge(df1,df2, how='left', on='key3') # Gives prefrence to the df1 as left dataframe and value matches to df2 or NaN
```

```python
pd.merge(df1,df2, how='right', on='key3') # Gives prefrence to the df2 as right dataframe and value matches to df1 or NaN
```

```python
pd.merge(df1, df2, how="outer", on='key3') # Gives preference to both right and left
```

### What if key names are different 

```python
import pandas as pd
```

```python
d11 = {'a':[1,2,3,4,5,"Omkar"],       
      'b' : [7,8,9,10,11,12],
      'c':[13,14,15,16,17,18],
      'd':[19,20,21,22,23,24]
}
d12 = {'key1':[1,"Shiva",3,4,5,"Omkar"],       
      'key2' : [7,8,9,10,11,12],
      'key3':[13,"Data Sci",15,"AI",17,18],
      'key4':[19,20,21,22,23,24]
}
df11 = pd.DataFrame(d11, index=[1,2,3,4,5,6])
df12 = pd.DataFrame(d12, index=[1, 2, 3, 4, 'v', 'vi'])
```

```python
df11
```

```python
df12
```

In above case we have different key names, Lets see how we can merge them. 

```python
# pd.merge(df11, df12, how="left", on='key1') # Gives key error
pd.merge(df11, df12, how="left", left_on='a', right_on= "key1") # we specify which key should be paired with which key 
```
